```python
nflx_originals = detail_df.copy()
nflx_originals['Premiere Date'] = nflx_originals.original_air_date.str.extract(r'(.*) \(')
nflx_originals['Premiere Date'] = nflx_originals['Premiere Date'].fillna(nflx_originals.original_air_date)
nflx_originals['Premiere Date'] = pd.to_datetime(nflx_originals['Premiere Date'])
nflx_originals = nflx_originals[nflx_originals['Premiere Date'].notna()]
nflx_originals['Month_Year'] = nflx_originals['Premiere Date'].map(lambda x: x.strftime('%b-%y'))

nflx_originals.head(30)
```

```python
# create month-year range
start_date = nflx_originals['Premiere Date'].min()
end_date = nflx_originals['Premiere Date'].max()
all_months =  pd.date_range(start_date,end_date,freq='MS').strftime('%b-%y')
all_months_df = pd.DataFrame(0, index = all_months,columns=["values"])
all_months_df.head()
```

```python
month_year_agg = nflx_originals.groupby(["Month_Year"]).agg(total_runtime = ('runtime',np.sum),
                                                            mean_rating = ('rating',np.mean),
                                                            median_rating = ('rating',np.median))
month_year_agg
```

```python
month_year_runtime = nflx_originals.groupby(["Month_Year"]).runtime.sum().rename("Month_Year")
month_year_runtime.head()
```

```python
all_months_df["values"] = all_months_df.merge(right = month_year_runtime,how = 'left', 
                                      left_index = True, right_index = True)["Month_Year"].fillna(0)
all_months_df.head()
```

```python
month_year_categories = all_months_df.index.values
# select values for display so we wont have too much overlap
month_year_categories_display = month_year_categories.copy()
i = 0
for monthYear in month_year_categories_display:
    if ("Jan" not in monthYear):
        monthYear = ""
    month_year_categories_display[i] = monthYear
    i+=1
# Build dataframe for month by month column chart
all_months_df.index = pd.CategoricalIndex(all_months_df.index, 
                               categories=month_year_categories, 
                               ordered=True)
all_months_df = all_months_df.sort_index()
# Take Min of content and make Hours
all_months_df["values_hr"] = all_months_df["values"]/60
```

## Burn Down Chart

```python
start_date = daily_sum_7.index.min()
end_date = daily_sum_7.index.max()
all_days =  pd.date_range(start_date,end_date)
all_days_df = pd.DataFrame(-60, index = all_days,columns=["1hpd_burn_rate"])
all_days_df['2hpd_burn_rate'] = -120
all_days_df['3hpd_burn_rate'] = -180
on_months_4x2 = ['May','Jun','Nov','Dec']
on_months_3x3 = ['Apr','May','Jun','Oct','Nov','Dec']
on_months_2x4 = ['Mar','Apr','May','Jun','Sep','Oct','Nov','Dec']
on_months_1x5 = ['Feb','Mar','Apr','May','Jun','Aug','Sep','Oct','Nov','Dec']

all_days_df['2hpd_burn_rate_2off_4on'] = np.where(all_days_df.index.strftime('%b').isin(on_months_2x4), -120, 0)
all_days_df['2hpd_burn_rate_1off_5on'] = np.where(all_days_df.index.strftime('%b').isin(on_months_1x5), -120, 0)
all_days_df['3hpd_burn_rate_3off_3on'] = np.where(all_days_df.index.strftime('%b').isin(on_months_3x3), -180, 0)
all_days_df['3hpd_burn_rate_2off_4on'] = np.where(all_days_df.index.strftime('%b').isin(on_months_2x4), -180, 0)
all_days_df['4hpd_burn_rate_4off_2on'] = np.where(all_days_df.index.strftime('%b').isin(on_months_4x2), -240, 0)
all_days_df['4hpd_burn_rate_3off_3on'] = np.where(all_days_df.index.strftime('%b').isin(on_months_3x3), -240, 0)

burn_down_df = all_days_df.merge(right = daily_sum_7,how = 'left', left_index = True,
                  right_index = True)[['1hpd_burn_rate','2hpd_burn_rate',
                                       '3hpd_burn_rate',
                                       '3hpd_burn_rate_3off_3on','3hpd_burn_rate_2off_4on',
                                       '2hpd_burn_rate_2off_4on','2hpd_burn_rate_1off_5on',
                                       '4hpd_burn_rate_4off_2on','4hpd_burn_rate_3off_3on',
                                       'runtime']].fillna(0).astype(int)
burn_down_df
```

```python
def adjusted_cumulative_sum(df,burn_rate_col,runtime_col):
    lastvalue = 0
    adjusted_cumulative_sum = []
    for index, row in df.iterrows():
        thisvalue = row[burn_rate_col] + row[runtime_col] + lastvalue
        if thisvalue < 0:
            thisvalue = 0
        adjusted_cumulative_sum.append(thisvalue)
        lastvalue = thisvalue
    return pd.Series(adjusted_cumulative_sum, index=df.index)

burn_down_df['1hpd_cum_sum'] = adjusted_cumulative_sum(burn_down_df,'1hpd_burn_rate','runtime')
burn_down_df['2hpd_cum_sum'] = adjusted_cumulative_sum(burn_down_df,'2hpd_burn_rate','runtime')
burn_down_df['3hpd_cum_sum'] = adjusted_cumulative_sum(burn_down_df,'3hpd_burn_rate','runtime')

burn_down_df['2hpd_cum_sum_2off_4on'] = adjusted_cumulative_sum(burn_down_df,'2hpd_burn_rate_2off_4on','runtime')
burn_down_df['2hpd_cum_sum_1off_5on'] = adjusted_cumulative_sum(burn_down_df,'2hpd_burn_rate_1off_5on','runtime')
burn_down_df['3hpd_cum_sum_3off_3on'] = adjusted_cumulative_sum(burn_down_df,'3hpd_burn_rate_3off_3on','runtime')
burn_down_df['3hpd_cum_sum_2off_4on'] = adjusted_cumulative_sum(burn_down_df,'3hpd_burn_rate_2off_4on','runtime')
burn_down_df['4hpd_cum_sum_4off_2on'] = adjusted_cumulative_sum(burn_down_df,'4hpd_burn_rate_4off_2on','runtime')
burn_down_df['4hpd_cum_sum_3off_3on'] = adjusted_cumulative_sum(burn_down_df,'4hpd_burn_rate_3off_3on','runtime')
burn_down_df
```
